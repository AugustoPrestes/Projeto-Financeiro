# Data Prep
Comos dados foram baixado mês a mês é necessário juntar todos os 12 arquivos em um único df

```python
# Criando a função para buscar e concatenar mais de um aquivo em um unico dataset

def load_datas(path, extension):
    # Salvando os arquivos em uma lista
    files = [file for file in os.listdir(path) if file.endswith(extension)]

    # Concatenando os arquivos em um DF
    dfs = []
    for file in files:
        d = pd.read_csv(os.path.join(path, file))
        dfs.append(d)

    return pd.concat(dfs, ignore_index=True)
```

```python
# Caminho da pasta
path = '/content/drive/MyDrive/AUGUSTO/augusto/dados/dados_financeiros/'
extension = '.csv'
```

```python
# Carregando o dataframe
df = load_datas(path, extension)
df.head()
```

```python
# Ajustando o indice do df
df.set_index('Identificador', inplace=True)
```

Com o nosso identificador ajustado como o indice dor registros, vamos criar novas colunas para facilitar a analise temporal

```python
# Adicionando colunas de mês
df['Mes'] = pd.to_datetime(df['Data'], dayfirst=True).dt.month     # Aqui será adicionado o número referente ao mês
df['Mes Nome'] = pd.to_datetime(df['Data'], dayfirst=True).dt.strftime("%b")   # Aqui será adicionado o nome referente ao mês

df.head()
```

```python
# Separando o tipo de transação
df['Tipo'] = df['Descrição'].str.split(' - ').str[0]
df.head()
```

```python
# Criando a função para pegar o nome do beneficiario
def split_text(t):
    if 'recebida' not in t:
        try:
            text = t.split('-')[1]
        except:
            pass
        else:
            return text
    else:
        return None


# Criando uma coluna para identificar se o registro é referente a uma receita ou despesa
def income_expense(valor):
    if valor < 0:
        return 'despesa'

    else:
        return 'receita'
```

```python
# Criando as colunas para identificar se o registro é um gasto ou uma receita
# e uma coluna com o nome do beneficiario
df['Beneficiario'] = df['Descrição'].apply(split_text)

df['Despesa/Receita'] = df['Valor'].apply(income_expense)
df.head()
```

```python
data = {'Hora_inicio': ['08:00', '10:30', '14:45'],
        'Hora_fim': ['09:30', '12:15', '16:30']}
df = pd.DataFrame(data)

# Converter as colunas para o tipo datetime
df['Hora_inicio'] = pd.to_datetime(df['Hora_inicio'])
df['Hora_fim'] = pd.to_datetime(df['Hora_fim'])

# Calcular a diferença entre a hora de fim e a hora de início para obter a duração
df['Duracao'] = df['Hora_fim'] - df['Hora_inicio']

# Calcular as horas acumuladas
df['Horas_acumuladas'] = df['Duracao'].cumsum()

print(df)
```
